Dados Ausentes

```python
dic = {'A' : [1, 2, np.nan], 'B': [5,np.nan, np.nan], 'C': [9, 10, 11] }
dic
```

```python
data = pd.DataFrame(dic)
data
```

```python
data.dropna()
```

```python
data.fillna(value = 0) #substitui os NaN por um valor específico, no caso value = 0
```

```python
data['A'].fillna(value = data['A'].mean()) #substitui os NaN por uma média, no caso .mean()
```

```python
data.fillna(method = 'ffill')
```

GroupBy

```python
dados = {'Empresa':['CESAR', 'BB','BB', 'CAIXA', 'CESAR', 'CAIXA' ],
       'Nome':['Everton', 'Carla','Clara', 'Raul', 'Hugo', 'Maria'],
       'Venda': [200, 320, 140, 125, 900,1200 ]}
```

```python
df = pd.DataFrame(dados)
df
```

```python
group = df.groupby('Empresa')
```

```python
group.sum()
```

```python
group.mean()
```

```python
group = df.groupby('Nome')
```

```python
group.sum()
```

Concatenar

```python
df1 = pd.DataFrame({'A':['A0','A1', 'A2', 'A3'],
                   'B':['B0','B1', 'B2', 'B3'],
                   'C':['C0','C1', 'C2', 'C3'],
                   'D':['D0','D1', 'D2', 'D3']},
                  index=[0,1,2,3])
df1
```

```python
df2 = pd.DataFrame({'A':['A4','A5', 'A6', 'A7'],
                   'B':['B4','B5', 'B6', 'B7'],
                   'C':['C4','C5', 'C6', 'C7'],
                   'D':['D4','D5', 'D6', 'D7']},
                  index=[4,5,6,7])
df2
```

```python
df3 = pd.DataFrame({'A':['A8','A9', 'A10', 'A11'],
                   'B':['B8','B9', 'B10', 'B11'],
                   'C':['C8','C9', 'C10', 'C11'],
                   'D':['D8','D9', 'D10', 'D11']},
                  index=[8,9,10,11])
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3], axis = 1)
```

Mesclar

```python
esquerda = pd.DataFrame({'key':['K0','K1', 'K2', 'K3'],
                   'A':['A0','A1', 'A2', 'A3'],
                   'B':['B0','B1', 'B2', 'B3']})
direita = pd.DataFrame({'key':['K0','K1', 'K2', 'K3'],
                   'C':['C0','C1', 'C2', 'C3'],
                   'D':['D0','D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita,how='inner',  on='key')
```
